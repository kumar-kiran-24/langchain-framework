```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
data=r"DataIngestion\speech.txt"
loader=TextLoader(data)
documents=loader.load()
documents
```

```python
texet_splitter=CharacterTextSplitter(chunk_size=100,chunk_overlap=10)
docs=texet_splitter.split_documents(documents)
docs[0]
docs[1]
```

```python
embeddings=OllamaEmbeddings(model="nomic-embed-text")
DB=FAISS.from_documents(docs,embeddings)
```

querying


```python
query="f Clairvaux. He was satirized in the 11th-century work, The Treatise of Garcia of Toledo, said to have been written by one of his canons."
docs=DB.similarity_search(query)
docs[0].page_content
```

# retriver

```python
retriver=DB.as_retriever()
rt1=retriver.invoke(query)
rt1[0]
```

### similarity search

```python
docs_and_sscore=DB.similarity_search_with_score(query)
```

```python
docs_and_sscore
```

```python
DB.save_local("fass_index")
```

```python
new_db=FAISS.load_local("fass_index",embeddings,allow_dangerous_deserialization=True)
```
