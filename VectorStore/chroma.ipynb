```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python

data=r"DataIngestion/speech.txt"
loader=TextLoader(data)
documents=loader.load()
```

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=10)
splits=text_splitter.split_documents(documents)
```

```python
embeddings=OllamaEmbeddings(model="nomic-embed-text")
DB=Chroma.from_documents(embedding=embeddings,documents=splits)
```

```python
query="f Clairvaux. He was satirized in the 11th-century work, The Treatise of Garcia of Toledo, said to have been written by one of his canons."
a1=DB.similarity_search_with_score(query=query)
```

```python
a2=DB.similarity_search(query)
a2[0].page_content
```
