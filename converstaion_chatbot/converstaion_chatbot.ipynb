```python
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage,AIMessage
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory,InMemoryChatMessageHistory
```

```python
from dotenv import load_dotenv
import os
api=os.getenv("GROQ_API_KEY")
```

```python
store={}

def get_session_history(sesion_id:str)->BaseChatMessageHistory:
    if sesion_id not in store:
        store[sesion_id]=InMemoryChatMessageHistory()
    return store[sesion_id]

```

```python
model=ChatGroq(model="openai/gpt-oss-20b",api_key=api)
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful chatbot. You must answer based on previous conversation."),
    MessagesPlaceholder(variable_name="history"),
    ("human", "{question}")
])
```

```python
chain=prompt|model
chabot=RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="question",
    history_messages_key="history"
)
```

```python
def ask_chatbot(user_message,session_id="user1"):
    response=chabot.invoke({"question":user_message},
                           config={"configurable":{"session_id":session_id}}
                           )
    return response
```

```python
def reset_memory(sesion_id="user1"):
    if sesion_id in store:
        store[sesion_id].clear()
    return "converstaion history cleared"
```

```python
print(ask_chatbot(user_message="wahat is ai",session_id="kiran"))
```
