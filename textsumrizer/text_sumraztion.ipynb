## Promt tepmlate text sumarization

```python
from langchain_core.prompts import PromptTemplate

generic_template = """
Write a summary of the following speech:

Speech: {speech}

Translate the precise summary to {language}.
"""

prompt = PromptTemplate(
    input_variables=["speech", "language"],
    template=generic_template
)
```

```python
complete_promt=prompt.format(language="kannada",speech=speech)
complete_promt
```

```python
llm.get_num_tokens(complete_promt)
```

```python
from langchain_classic.chains import LLMChain
lm_chain=llm|prompt
```

```python
lm_chain = LLMChain(
    llm=llm,
    prompt=prompt
)
```

```python
response=lm_chain.run({
    "speech":speech,
    "language":"Kannada"
})
```
